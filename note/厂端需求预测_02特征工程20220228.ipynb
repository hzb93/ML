## 根据平均销量创建abc特征
你这样分一下类：①所有零件销量加起来，②各零件按销量倒序排序，③前n1零件组成销量占比50%，④第n1 - n2组成销量占比额外20%（累计70%），⑤第n2 - n3组成销量占比额外10%（累计80%），⑥第n3 - n4组成销量占比额外10%（累计90%），⑦第n4 - n5组成销量占比额外5%（累计95%），⑧第n5 - n6组成销量占比额外5%（累计100%）。 零件-仓库算一个记录 用样本内销量平均值来做吧，N+0、N+1都用样本内的分类

```python
# 根据月均值 purchased_quantity_sum_mean 来判别 销量类型 abc
# 输入的是df
def make_abc(df):
    df = df.sort_values(["storage_code","purchased_quantity_sum_mean"],ascending=[True,False])
    df_ = df.groupby("storage_code")[["purchased_quantity_sum_mean"]].sum()
    df_.reset_index()
    df = df.merge(df_,on="storage_code",how="left")
    df["purchased_quantity_ratio"] = df["purchased_quantity_sum_mean_x"]/df["purchased_quantity_sum_mean_y"]
    df_cumsum = df.groupby("storage_code").apply(lambda x:x.purchased_quantity_ratio.cumsum())
    df_cumsum = pd.DataFrame(df_cumsum).reset_index(drop=True)
    df_cumsum.columns = ["purchased_quantity_ratio_cumsum"]
    df = pd.concat([df,df_cumsum],axis=1)
    return df
```

```python
df_holiday = pd.DataFrame()
for i in [2019,2020,2021,2022]:
    temp = pd.DataFrame(cal.holidays(i))
    df_holiday = df_holiday.append(temp)
df_holiday.rename(columns={0:"month",1:"holiday"},inplace=True)
```

```python
df_holiday.month = pd.to_datetime(df_holiday.month,format='%Y-%m-%d')
```

```python
df = df.merge(df_holiday,on="month",how="left")
df.fillna(value="not holiday",inplace=True)
```

```python
def count_other(x):
    if (x in ["Chinese New Year's eve",'Chinese New Year','Spring Festival']) or (x=="not holiday") :
        return 0
    else:
        return 1
```

```python
def count_spring(x):
    if x in ["Chinese New Year's eve",'Chinese New Year','Spring Festival']:
        return 1
    else:
        return 0
```

```python
df.spring_count = df.holiday.apply(lambda x:count_spring(x))
```

```python
df.other_count = df.holiday.apply(lambda x:count_other(x))
```

```python
df = df.set_index("month")[["spring_count","other_count"]].resample("M").sum().reset_index()
```

```python
# df['date'] = df['date'].dt.to_period('M')
df['month'] = df['month'].apply(lambda x:x.strftime("%Y-%m"))
df
```
